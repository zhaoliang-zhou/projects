## Rule: 
In the game Trouble, the goal is to move a bunch of markers from a starting point all the way to the ending point. Ignoring some subtleties of the game, the basic move is very simple:

You roll a standard 6-sided die.
You move your marker the number of spaces equal to the value on the die.
If you roll a six, then you roll and move again. You keep doing this as long as you roll sixes.
For example, if you roll a 3, then you move 3 spaces. If you roll a 6 followed by a 3, then you move 6 + 3 = 9 spaces. If you roll a 6, another 6, and then a 3, then you move 6 + 6 + 3 = 15 spaces.

#"Go again on 6"

```
import random
def move6():
  roll = random.randint(1,6)
  total = roll
  while roll == 6:                  ## if the dice shows 6, roll again
    roll=random.randint(1,6)
    total += roll 
  return total 
  
```

#Scenario 

You are playing a head-to-head game with another person. You use the go again on 6 rule,
the other person uses the go again on 1 rule. You both roll your die to complete a move.
The winner is who ever has a the larger move. (If there is a tie, repeat until there is a winner.)
Is this a fair game, in the sense that both players have an equal probability of winning? If
not, who has the advantage, and what is the corresponding probability?


For this scenerio, I would like to investigate using hypothesis test

#Hypothesis Test

For the scenario above, to investigate whether it is fair to play "roll again on 1" against "roll again on 6", I would actlly like to investigate whether there is a difference in means for the moves "roll again on 6" and "roll again ". If there is no difference or the difference is not statistically significant between the averge length of moves using two rules, then we could say that this game is fair. On the other hand, if there is a difference in average move length, we could conclude that this is not a fair game; one has more chance of winning.

Denote $\hat{x}1$ to the average length of move using "roll again on 1" rule.

Denote $\hat{x}6$ to the average length of move using "roll again on 6" rule

Here is my hypothesis test:

Null hypothese:   $\hat{x}1$ = $\hat{x}6$. There is no difference in average length of move between two rules. This is a fair game.

Alternative hypothesis: $\hat{x}1$ != $\hat{x}6$. There is a difference in average length of move between two rules. This is game is not fair

Since both are not normally distributed, so we could not use t-test or student t-test since the asumptions of normality (normal distribution) and equal variance are violated. Thus, we would use Welch's t-test, which does not make those assumpstions. 

For Welch's t-test: 


t = (x1-x2)/sqrt( (s1^2 / n1)   + ( s2^2 / n2 ) )

where x1 and x2 are sample mean, s1 and s2 are sample varianc, and n1 and n2 are sample size

For the formula:

Welch's t-test: https://en.wikipedia.org/wiki/Welch%27s_t-test

Degree of freedom for Welch's t-test: https://stats.stackexchange.com/questions/48636/are-the-degrees-of-freedom-for-welchs-test-always-less-than-the-df-of-the-poole

"The Welch test uses the Satterthaite-Welch adjustment for the degrees of freedom:"

Next, write a function to compute Welch's t-test and degree of freedom. These are functions to calculate t-statistics and 2 samples and degree of freedom

```
def welch_t(a, b): ## here a and b are just two sample: sample a and sample b
    diff = np.mean(a) - np.mean(b)
    var = np.sqrt(np.var(a)/np.size(a) + np.var(b)/np.size(b))
    return diff/var
```

```
def welch_df(a, b): ## Calculate the effective degrees of freedom for two samples.
    s_a, s_b = np.var(a), np.var(b)
    n_a, n_b = np.size(a), np.size(b)
    numer = (s_a/n_a + s_b/n_b)**2
    denom = (s_a/ n_a)**2/(n_a - 1) + (s_b/ n_b)**2/(n_b - 1)
    return numer/denom
```

After writting the functions, now actually calculate the t-statistic and df

```
t = welch_t(List6, List1)
df = welch_df(List6, List1)

```

Here we have a very small t-statistics, and what does it look like on a normal t-distribution. 

```
fig, ax = plt.subplots()
x = np.linspace(-4, 4, 200)
t_dist = stats.t.pdf(x, df)
ax.plot(x, t_dist, label='t-distribution')
ax.vlines(t, 0, 0.4, linestyles='--', color='k', 
          label='t-statistic = {:.3f}'.format(t))
ax.set_ylabel('pdf(t)')
ax.set_xlabel('t')
ax.set_ylim(0, 0.55)
ax.legend(loc=9)
ax.fill_between(x, 0, t_dist, where=(x >= t), color='grey', alpha=0.5)
```

We saw a very small t-statistics, and on the graph above, it is very close to 0 in a normal t distribution, but this might not be so convincing, let us try to calculate the p-value corresponding the df and t-statistics. 

```
def p_value(a, b):
    "Calculate two-sided t-test for two samples a and b."
    t = welch_t(b, a)
    df = welch_df(b, a)
    p = stats.t.cdf(-np.abs(t), df)
    return t, 2*p
```

```
p_value(List6,List1)
```

 For P value  in this case, it is the probability of obtaining a difference in means between moves of "roll again on 6" and "roll again on 1"at least as extreme as the one in my sample data, assuming the truth of the null hypothesis. In this case, the p-value is 0.94. Using the standard cutoff 0.05, the p-value is bigger than 0.05, thus we would fail to reject the null hypothesis: there is no statistically siginificant difference in average moves of "roll again on 6" and "roll again on 1". Thus, it seems like this game is fair.
